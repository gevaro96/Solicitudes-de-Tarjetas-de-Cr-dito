# Cuales son los tres clientes solteros, con coche y con 30-59 días de atraso que tienen los mayores ingresos


```python
##primero hacemos el join con el mes mas reciente
c.execute('''
    SELECT a.ID, NAME_FAMILY_STATUS, FLAG_OWN_CAR, MONTHS_BALANCE, STATUS, AMT_INCOME_TOTAL 
    FROM application_record as a
    LEFT JOIN credit_record as c
    ON a.ID=c.ID
    WHERE NAME_FAMILY_STATUS ='Single / not married' 
        AND FLAG_OWN_CAR = 'Y'
        AND MONTHS_BALANCE=0
        AND STATUS= 1
    ORDER by AMT_INCOME_TOTAL desc
    LIMIT 3
    ''').fetchall() 
```

# Ahora, una pequeña limpieza de datos

# Ahora definimos los diccionarios

```python
#creamos el diccionario para la variable NAME_INCOME_TYPE
c.execute('''
    CREATE TABLE IF NOT EXISTS NAME_INCOME_TYPE_Dicc AS
        SELECT DISTINCT NAME_INCOME_TYPE, ROW_NUMBER() OVER() as NAME_INCOME_TYPE_INT
        FROM application_record
        GROUP BY NAME_INCOME_TYPE
        ORDER BY NAME_INCOME_TYPE
    ''').fetchall()
#creamos el diccionario para la variable NAME_EDUCATION_TYPE
c.execute('''
    CREATE TABLE IF NOT EXISTS NAME_EDUCATION_TYPE_Dicc AS
        SELECT DISTINCT NAME_EDUCATION_TYPE, ROW_NUMBER() OVER() AS NAME_EDUCATION_TYPE_INT
        FROM application_record
        GROUP BY NAME_EDUCATION_TYPE
        ORDER BY NAME_EDUCATION_TYPE
    ''').fetchall()
#creamos el diccionario para la variable NAME_FAMILY_STATUS
c.execute('''
    CREATE TABLE IF NOT EXISTS NAME_FAMILY_STATUS_Dicc AS
        SELECT DISTINCT NAME_FAMILY_STATUS, ROW_NUMBER() OVER() AS NAME_FAMILY_STATUS_INT
        FROM application_record
        GROUP BY NAME_FAMILY_STATUS
        ORDER BY NAME_FAMILY_STATUS
    ''').fetchall()
#creamos el diccionario para la variable NAME_HOUSING_TYPE
c.execute('''
    CREATE TABLE IF NOT EXISTS NAME_HOUSING_TYPE_Dicc AS
        SELECT DISTINCT NAME_HOUSING_TYPE, ROW_NUMBER() OVER() AS NAME_HOUSING_TYPE_INT
        FROM application_record
        GROUP BY NAME_HOUSING_TYPE
        ORDER BY NAME_HOUSING_TYPE
    ''').fetchall()
#creamos el diccionario para la variable OCCUPATION_TYPE
c.execute('''
    CREATE TABLE IF NOT EXISTS OCCUPATION_TYPE_Dicc AS
        SELECT DISTINCT OCCUPATION_TYPE, ROW_NUMBER() OVER() AS OCCUPATION_TYPE_INT
        FROM application_record
        WHERE OCCUPATION_TYPE IS NOT NULL
        GROUP BY OCCUPATION_TYPE
        ORDER BY OCCUPATION_TYPE
    ''').fetchall()
#notamos que se asignó el numero 1 a los campos 
#Ahora guardamos estos diccionarios como CSV para poder compartir los diccionarios junto con los nuevos datos 
#actualizados con informacion categórica numérica

##Guardamos el diccionario para NAME_INCOME_TYPE
db_df = pd.read_sql_query("SELECT * FROM NAME_INCOME_TYPE_Dicc", conn)
db_df.to_csv('NAME_INCOME_TYPE_Dicc.csv', index=False)

##Guardamos el diccionario para NAME_EDUCATION_TYPE
db_df = pd.read_sql_query("SELECT * FROM NAME_EDUCATION_TYPE_Dicc", conn)
db_df.to_csv('NAME_EDUCATION_TYPE_Dicc.csv', index=False)

##Guardamos el diccionario para NAME_FAMILY_STATUS
db_df = pd.read_sql_query("SELECT * FROM NAME_FAMILY_STATUS_Dicc", conn)
db_df.to_csv('NAME_FAMILY_STATUS_Dicc.csv', index=False)

##Guardamos el diccionario para NAME_HOUSING_TYPE
db_df = pd.read_sql_query("SELECT * FROM NAME_HOUSING_TYPE_Dicc", conn)
db_df.to_csv('NAME_HOUSING_TYPE_Dicc.csv', index=False)

##Guardamos el diccionario para OCCUPATION_TYPE
db_df = pd.read_sql_query("SELECT * FROM OCCUPATION_TYPE_Dicc", conn)
db_df.to_csv('OCCUPATION_TYPE_Dicc.csv', index=False)
```

# Creación y formateo de nueva tabla application_record_int
La siguiente tabla contiene la misma información que la original pero se hacen los cambios en las siguientes coulumnas categóricas  
* CODE_GENDER  
* FLAG_OWN_CAR  
* FLAG_OWN_REALTY  
* NAME_INCOME_TYPE  
* NAME_EDUCATION_TYPE  
* NAME_FAMILY_STATUS  
* NAME_HOUSING_TYPE  
* OCCUPATION_TYPE    

Por sus correspondientes columnas en enteros conforme a los diccionarios previamente creados.


```python
application_record.to_sql('application_record_int', conn, if_exists='replace', index=False)
c.execute('''
    UPDATE application_record_int
    SET CODE_GENDER = CASE WHEN CODE_GENDER ='M' THEN 1 ELSE 2 END,
        FLAG_OWN_CAR = CASE WHEN FLAG_OWN_CAR ='Y' THEN 1 ELSE 2 END,
        FLAG_OWN_REALTY = CASE WHEN FLAG_OWN_REALTY ='Y' THEN 1 ELSE 2 END,
        NAME_INCOME_TYPE = (SELECT NAME_INCOME_TYPE_INT FROM NAME_INCOME_TYPE_Dicc WHERE 
                application_record_int.NAME_INCOME_TYPE=NAME_INCOME_TYPE_Dicc.NAME_INCOME_TYPE),
        NAME_EDUCATION_TYPE = (SELECT NAME_EDUCATION_TYPE_INT FROM NAME_EDUCATION_TYPE_Dicc WHERE 
                application_record_int.NAME_EDUCATION_TYPE=NAME_EDUCATION_TYPE_Dicc.NAME_EDUCATION_TYPE),
        NAME_FAMILY_STATUS = (SELECT NAME_FAMILY_STATUS_INT FROM NAME_FAMILY_STATUS_Dicc WHERE 
                application_record_int.NAME_FAMILY_STATUS=NAME_FAMILY_STATUS_Dicc.NAME_FAMILY_STATUS),
        NAME_HOUSING_TYPE = (SELECT NAME_HOUSING_TYPE_INT FROM NAME_HOUSING_TYPE_Dicc WHERE 
                application_record_int.NAME_HOUSING_TYPE=NAME_HOUSING_TYPE_Dicc.NAME_HOUSING_TYPE),
        OCCUPATION_TYPE = (SELECT OCCUPATION_TYPE_INT FROM OCCUPATION_TYPE_Dicc WHERE 
                application_record_int.OCCUPATION_TYPE=OCCUPATION_TYPE_Dicc.OCCUPATION_TYPE)
    ''').fetchall() 
#Ahora simplemente renombramos las culumnas que modificamos para señalizar que solamente hicimos un pequeño cambio de nom,bre
c.execute('''
    ALTER TABLE application_record_int
    RENAME COLUMN CODE_GENDER TO CODE_GENDER_INT
    ''').fetchall() 
c.execute('''
    ALTER TABLE application_record_int
    RENAME COLUMN FLAG_OWN_CAR TO FLAG_OWN_CAR_INT
    ''').fetchall()
c.execute('''
    ALTER TABLE application_record_int
    RENAME COLUMN FLAG_OWN_REALTY TO FLAG_OWN_REALTY_INT
    ''').fetchall()
c.execute('''
    ALTER TABLE application_record_int
    RENAME COLUMN NAME_INCOME_TYPE TO NAME_INCOME_TYPE_INT
    ''').fetchall()
c.execute('''
    ALTER TABLE application_record_int
    RENAME COLUMN NAME_EDUCATION_TYPE TO AME_EDUCATION_TYPE_INT
    ''').fetchall()
c.execute('''
    ALTER TABLE application_record_int
    RENAME COLUMN NAME_FAMILY_STATUS TO NAME_FAMILY_STATUS_INT
    ''').fetchall()
c.execute('''
    ALTER TABLE application_record_int
    RENAME COLUMN NAME_HOUSING_TYPE TO NAME_HOUSING_TYPE_INT
    ''').fetchall()
c.execute('''
    ALTER TABLE application_record_int
    RENAME COLUMN OCCUPATION_TYPE TO OCCUPATION_TYPE_INT
    ''').fetchall()
```
